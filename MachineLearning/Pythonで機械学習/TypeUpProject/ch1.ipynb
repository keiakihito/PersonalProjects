```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Random number
D = 100
N = 2
xdata = np.random.rand(D,N).reshape(D,N).astype(np.float32)
```

```python
print(xdata)
```

```python
plt.scatter(xdata[:,0],xdata[:,1])
plt.show()
```

```python
def f(x):
    y = x * x
    return y
```

```python
tdata = (xdata[:,1] > f(xdata[:,0])).astype(np.int32)
```

```python
print(tdata)
```

```python
print(tdata.shape)
```

```python
ndata0 = np.where(tdata == 0)
ndata1 = np.where(tdata == 1)
```

```python
print(ndata1)
```

```python
print(ndata0)
```

```python
x = np.linspace(-2.0, 2.0, D)
plt.plot(x,f(x))
plt.scatter(xdata[ndata0,0], xdata[ndata0,1], marker = "x")
plt.scatter(xdata[ndata1,0], xdata[ndata1,1], marker = "o")
plt.show()
```
